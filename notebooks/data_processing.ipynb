```python
# Data Processing 
# Data Loading & Structure Analysis
# Import necessary libraries #Make sure pandas is installed in environment
import pandas as pd

# Load the datasets
train = pd.read_csv('../data/train.csv')
test = pd.read_csv('../data/test.csv')
submission = pd.read_csv('../data/submission.csv')


```

```python
# Quick peek at each dataset
print("Train Dataset Preview:")
display(train.head())

print("\nTest Dataset Preview:")
display(test.head())

print("\nSubmission File Preview:")
display(submission.head())
```

```python
#Structure Analysis
# 1. Dataset Dimensions
print(f"Train dataset: {train.shape[0]} rows × {train.shape[1]} columns")
print(f"Test dataset: {test.shape[0]} rows × {test.shape[1]} columns")

# 2. Data types
print("\nData Types in Train dataset:")
print(train.dtypes)
```

```python
#Missing Values
print("\nMissing Values in Train dataset:")
missing_counts = train.isnull().sum()
missing_percentage = (missing_counts / len(train)) * 100

#Table of column(features) with missing variables from highest to lowest
missing_df = pd.DataFrame({'Missing Values': missing_counts, 'Percentage': missing_percentage})
display(missing_df[missing_df['Missing Values'] > 0].sort_values('Missing Values', ascending=False))
```

```python
#Types of features
categorical_features = train.select_dtypes(include=['object']).columns.tolist()
numerical_features = train.select_dtypes(include=['number']).columns.tolist()

print(f"\nNumber of Categorical Features: {len(categorical_features)}")
print(f"Categorical Features: {categorical_features}")

print(f"\nNumber of Numerical Features: {len(numerical_features)}")
print(f"Numerical Features: {numerical_features}")
```
